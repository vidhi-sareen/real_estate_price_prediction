```python
#convert sqft value to float
def float_con(x):
  try:
    float(x)
  except:
    return False
  return True
```

```python
df1[~df1['total_sqft'].apply(float_con)].head(15)
```

```python
## converting the range values to the average of those two values
def con_to_num(x):
  t = x.split('-')
  if len(t) == 2:
    return (float(t[0]) + float(t[1]))/2
  try:
    return float(x)
  except:
    return None
```

```python
df2 = df1.copy()
df2['total_sqft'] = df2['total_sqft'].apply(con_to_num)
df2.head(15)
```

Feature engineering

```python
# creating a new col 'price_per_sqft' and assigning the price value per sqft
df3 = df2.copy()
df3['price_per_sqft'] = (df3['price']*100000) / df3['total_sqft']
df3.head(15)
```
